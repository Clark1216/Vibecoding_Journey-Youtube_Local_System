This notebook shows a **beginner-friendly** way to download YouTube videos in **1080p or higher** using `yt-dlp`.

## 2) Import library

We import `YoutubeDL` from `yt_dlp`.

```python
from yt_dlp import YoutubeDL
```

## 3) Configure inputs

- `url`: YouTube link from user
- `output_template`: output file naming pattern
- `allow_lower_resolution`: if `False`, download fails when 1080p+ is unavailable

```python
url = input('Enter YouTube URL: ').strip()
output_template = '%(title)s.%(ext)s'
allow_lower_resolution = False  # Set to True to allow fallback below 1080p

if not url:
    raise ValueError('No URL provided. Please run the cell again and enter a URL.')
```

## 4) Build yt-dlp options

This format selector means:
1. Prefer MP4 video streams with `height>=1080` + M4A audio.
2. If not available, try any 1080p+ video + best audio.
3. If still not available, try a single 1080p+ combined stream.
4. Optional final fallback to `best` if enabled.

```python
preferred_hd = (
    'bestvideo[height>=1080][ext=mp4]+bestaudio[ext=m4a]/'
    'bestvideo[height>=1080]+bestaudio/best[height>=1080]'
)
format_selector = preferred_hd
if allow_lower_resolution:
    format_selector = f'{preferred_hd}/bestvideo+bestaudio/best'

ydl_opts = {
    'format': format_selector,
    'outtmpl': output_template,
    'merge_output_format': 'mp4',
    'noplaylist': True,
}

ydl_opts
```

## 5) Download video

If download fails, the error message should explain why (for example, no 1080p streams available).

```python
try:
    with YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])
    print('Download complete.')
except Exception as exc:
    print(f'Download failed ({exc.__class__.__name__}): {exc}')
```

## Learning tips for juniors

- Start by changing only one variable at a time (like `output_template`).
- Print and inspect `ydl_opts` before downloading to understand configuration.
- Test with short videos first.
- Keep error messages; they help debugging quickly.
